```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import os
import glob
import intake

mpl.rcParams['figure.figsize'] = 12, 8
mpl.rcParams['font.size'] = 18
```

```python
def load_data():
    ddir = '/glade/scratch/apauling/co2-cmip6/normal/'
    dfiles = glob.glob(ddir+'*.nc')
    
    models = set([f.split('/')[-1].split('_')[2] for f in dfiles])
    ds_dict = dict()
    for model in models:
        print(model)
        modelfiles = glob.glob(ddir+'*_'+model+'_*.nc')
    
        ds_dict[model] = xr.open_mfdataset(modelfiles, combine='by_coords')
    
    return ds_dict
```

```python
ds_dict = load_data()
```

```python
def compute_anom(ds_dict):
    
    data_dict = dict()
    
    for model in ds_dict.keys():
        print(model)
        ds = ds_dict[model]
        
        fgco2 = ds['fgco2'].squeeze()
        
        fgco2_h = fgco2.sel(time=slice('1981-01', '2000-12')).mean('time')
        fgco2_s = fgco2.sel(time=slice('2081-01', '2100-12')).mean('time')
        
        fgco2_anom = fgco2_s - fgco2_h
        if 'depth' in list(fgco2_anom.coords.keys()):
            fgco2_anom = fgco2_anom.drop('depth')
        
        if model == 'IPSL-CM6A-LR':
            fgco2_anom = fgco2_anom.rename({'nav_lat': 'latitude', 'nav_lon': 'longitude'})
        if model[:5] == 'CESM2' or model == 'CNRM-ESM2-1':
            print(list(fgco2.coords.keys()))
            fgco2_anom = fgco2_anom.rename({'lat': 'latitude', 'lon': 'longitude'})
        
        data_dict[model] = fgco2_anom
        
    return data_dict

data_dict = compute_anom(ds_dict)
```

```python
def plot_stereo(fig, axnum, data, title):
    limit = int(data.shape[0]/2)
    if title[:3] != 'MPI':
        dplt = data[:limit, :]
    else:
        dplt = data[limit:, :]
    axs = fig.add_subplot(5, 3, axnum, projection=ccrs.SouthPolarStereo())
    axs.coastlines()
    axs.gridlines()
    axs.set_extent([-180, 180, -90, -40], crs=ccrs.PlateCarree())
    im = dplt.plot.pcolormesh('longitude', 'latitude', ax=axs, transform=ccrs.PlateCarree(),
                                 cbar_kwargs={'label': 'kg m$^{-2}$ s$^{-1}$'})
        
    im.set_clim(-4e-9, 4e-9)
    axs.set_title(title)
    
    return im, axs
```

```python
fig = plt.figure(1, figsize=(18, 24))

for n, key in enumerate(data_dict.keys()):
    print(key)
    data = data_dict[key]
    print(list(data.coords.keys()))
   
    im, axs = plot_stereo(fig, n+1, data, key)
```

```python
fig.savefig('../figures/co2/fgco2_anomaly_normal.png',
           bbox_inches='tight')
```
