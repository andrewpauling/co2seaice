```python
import xarray as xr 
import os
from glob import glob
import utils as ut

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.util import add_cyclic_point
```

```python
def get_hist_co2(institute, model, expt, ens, vartype, var):
    droot = '/glade/collections/cmip/CMIP6/CMIP'
    ddir = os.path.join(droot, institute, model, expt, ens, vartype, var)
    if len(glob(ddir+'/*')) > 1:
        dfiles = glob(ddir+'/gn/latest/'+var+'_'+vartype+'_'+model+'_'+expt+'_'+ens+'_gn_*.nc')
    elif vartype is 'SImon': 
        dfiles = glob(ddir+'/gn/latest/'+var+'_'+vartype+'_'+model+'_'+expt+'_'+ens+'_gn_*.nc')
    else:
        dfiles = glob(ddir+'/gr/**/*/'+var+'_'+vartype+'_'+model+'_'+expt+'_'+ens+'_gr_*.nc')
            
    ds = xr.open_mfdataset(dfiles, combine='by_coords')
    
    return ds
```

```python
def find_lat_lon(ds):
    dimlist = [key for key in ds.dims]
    if 'nlat' in dimlist:
        dim_lat = 'nlat'
        dim_lon = 'nlon'
    else:
        dim_lat = 'lat'
        dim_lon = 'lon'
    return dim_lat, dim_lon 
```

```python
def area_weighted_sum(data,dim_lat,dim_lon):
    """
    Compute area weighted sum of some CMIP6 data.

    Parameters
    ----------
    data : xarray DataArray
        DataArray containing the data to average.
        Assumes lat dimension/coordinate is called
        'lat' and lon dimension/coordinate is called
        'lon'

    Returns
    -------
    dout : xarray DataArray
        DataArray containing area-weighted sum of
        data. Has lost domensions of lat and lon

    """
    
    lat = data.lat
    coslat = np.cos(np.deg2rad(lat))
    
    dout = (data*coslat).sum((dim_lat, dim_lon))
    
    return dout
```

```python
fgco2_data = get_hist_co2('NCAR', 'CESM2', 'esm-hist', 'r1i1p1f1','Omon','fgco2')
siconc_data = get_hist_co2('NCAR', 'CESM2', 'esm-hist', 'r1i1p1f1','SImon','siconc')
siarea_data = get_hist_co2('NCAR', 'CESM2', 'esm-hist', 'r1i1p1f1','SImon','siareas')
```

```python
fgco2 = fgco2_data['fgco2']
siconc = siconc_data['siconc']
siareas = siarea_data['siareas']
fgco2
```

```python
# Calculate seasonal means for each year: 
fgco2_rmn = fgco2.rolling(center=True, time=3).mean()
fgco2_seasonal = fgco2_rmn.isel(time = slice(None,None,3))

siconc_rmn = siconc.rolling(center=True, time=3).mean()
siconc_seasonal = siconc_rmn.isel(time = slice(None,None,3))

siarea_rmn = siareas.rolling(center=True, time=3).mean()
siarea_seasonal = siarea_rmn.isel(time = slice(None,None,3))
```

```python
# Calculate seasonal mean across all years: 
fgco2_seasonal_mn = fgco2.groupby('time.season').mean('time')
siconc_seasonal_mn = siconc.groupby('time.season').mean('time')
```

```python
sia_DJF = siarea_seasonal.isel(time=slice(None,None,4))
sia_MAM = siarea_seasonal.isel(time=slice(1,None,4))
sia_JJA = siarea_seasonal.isel(time=slice(2,None,4))
sia_SON = siarea_seasonal.isel(time=slice(3,None,4))
```

```python
fgco2_DJF = fgco2_seasonal.isel(time=slice(None,None,4))
fgco2_MAM = fgco2_seasonal.isel(time=slice(1,None,4))
fgco2_JJA = fgco2_seasonal.isel(time=slice(2,None,4))
fgco2_SON = fgco2_seasonal.isel(time=slice(3,None,4))
```

```python
[dim_lat, dim_lon] = find_lat_lon(fgco2_DJF)

fgco2_tot_DJF = area_weighted_sum(fgco2_DJF.where(fgco2_DJF.lat<-60.0), dim_lat, dim_lon)
fgco2_tot_MAM = area_weighted_sum(fgco2_MAM.where(fgco2_MAM.lat<-60.0), dim_lat, dim_lon)
fgco2_tot_JJA = area_weighted_sum(fgco2_JJA.where(fgco2_JJA.lat<-60.0), dim_lat, dim_lon)
fgco2_tot_SON = area_weighted_sum(fgco2_SON.where(fgco2_SON.lat<-60.0), dim_lat, dim_lon)
```

```python
years = fgco2_tot_DJF.time.dt.year
```

```python
plt.scatter(years,fgco2_tot_DJF)
plt.ylim(0,3e-06)
plt.title('Total CO2 flux: DJF')
```
